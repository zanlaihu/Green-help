Using read function to read the data, so that I can have a quick look.

```python
import pandas as pd

csv_file = "../input/ausnut1.csv"  # path to the downloaded csv file
ausdf = pd.read_csv(csv_file)
ausdf.head(5)
```

There are many useless columns in the dataset. Firstly, I will leave only useful columns, Which are 'Food ID',  'Food Name',  'Energy, with dietary fibre (kJ)',  'Protein (g)',  'Added sugars (g)'

```python
columns_to_keep = ['Food ID',
                    'Food Name',
                    'Energy, with dietary fibre (kJ)',
                    'Protein (g)',
                    'Added sugars (g)'
                  ]
    
newdf = ausdf[columns_to_keep]
newdf.head(5)
```

Add two columns based on the value of Energy and Protein/added sugar.

```python
newdf = newdf.assign(protein_ratio=newdf['Protein (g)'] / newdf['Energy, with dietary fibre (kJ)'])
newdf = newdf.assign(addedsugar_ratio=newdf['Added sugars (g)'] / newdf['Energy, with dietary fibre (kJ)'])
newdf    
```

I need to find the maximum value so that I can split them by different stars.

```python
by_protein = newdf.sort_values(by=["protein_ratio"], ascending=False)
by_protein.head(5)
```

```python
by_addsugar = newdf.sort_values(by=["addedsugar_ratio"], ascending=False)
by_addsugar.head(5)
```

Then I create two columns based on the protein ratio and added sugar ratio.

```python
    def function1(a):
        if a < 0.006:
            return 1
        elif a < 0.012:
            return 2
        elif a < 0.018:
            return 3
        elif a < 0.024:
            return 4
        elif a < 0.03:
            return 5
        elif a < 0.036:
            return 6
        elif a < 0.042:
            return 7
        elif a < 0.048:
            return 8
        elif a < 0.054:
            return 9
        else:
            return 10

    def function2(a):
        if a < 0.006:
            return 10
        elif a < 0.012:
            return 9
        elif a < 0.018:
            return 8
        elif a < 0.024:
            return 7
        elif a < 0.03:
            return 6
        elif a < 0.036:
            return 5
        elif a < 0.042:
            return 4
        elif a < 0.048:
            return 3
        elif a < 0.054:
            return 2
        else:
            return 1
newdf['protein_star'] = newdf.apply(lambda x: function1(x.protein_ratio), axis=1)
newdf['addedsugar_star'] = newdf.apply(lambda x: function2(x.addedsugar_ratio), axis=1)
newdf.head(5)
```

Finally, add a new column called aver_star, which is the average star of protein_star and addedsugar_star.

```python
def function3(a, b):
    c = (a + b)/2
    return c
        
newdf['aver_star'] = newdf.apply(lambda x: function3(x.protein_star, x.addedsugar_star), axis=1)
newdf.head(10)
```

Sort it from top to bottom. Here is the top 10 healthy food.
